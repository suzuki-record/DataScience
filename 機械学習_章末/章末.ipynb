```python
d_train = pd.read_csv('cancer_train.csv')
d_test = pd.read_csv('cancer_test.csv')
```

# Data Analysis

```python
d_train.head(3)
```

```python
d_test.head(3)
```

```python
x_train = d_train.drop('diagnosis', axis = 1).values
t_train = d_train['diagnosis'].values
x_test = d_test
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

```python
std_scaler.fit(x_train)
```

```python
# 標準化
x_train_std = std_scaler.transform(x_train)
x_test_std = std_scaler.transform(x_test)
```

# SVM

```python
# モデルの定義
from sklearn.svm import SVC
svc = SVC()
```

```python
# モデルの学習
svc.fit(x_train_std, t_train)
```

```python
# モデルの検証
print('train score : ', svc.score(x_train_std, t_train))
print('test score : ', svc.score(x_test_std, t_test))
```

```python
y_pred = svc.predict(x_test_std)
```

```python
# 推論結果を、 success, fail に書き換える
y_list = []
for y in y_pred:
  if y == 'B':
    f = y.replace("B", '0')
    y_list.append(f)
  else:
    s = y.replace("M", '1')
    y_list.append(s)
```

```python
pd.Series(y_list, name='pred').to_csv('submission.csv', index=None)
```
