# 欠損値処理

- 数値変数
  - 分布のばらつきが小さい・外れ値がない
    - 平均値で補完
  - 分布のばらつきが大きい・外れ値がある
    - 中央値で補完
- カテゴリカル変数
  - 最頻値で補完

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['price','horsepower','peak-rpm']) # デフォルトでaxis=0なので、行方向に削除が実行されます。
```

```python
df.shape
```

```python
df = df.drop(labels='normalized-losses', axis=1) # 欠損値が多い列は列ごと削除
```

```python
df.shape
```

```python
# 欠損値の確認
df.isnull().sum()
```

```python
df['bore'].describe()
```

```python
plt.hist(df['bore']);
```

```python
df['stroke'].describe()
```

```python
plt.hist(df['stroke']);
```

```python
# 欠損値を平均値で補完
df = df.fillna({'bore':df['bore'].mean(), 'stroke':df['stroke'].mean()})
```

```python
# 欠損値の確認
df.isnull().sum()
```

```python
df['num-of-doors'].unique()
```

```python
df['num-of-doors'].nunique()
```

```python
# 最頻値を使用して欠損値を補完
df = df.fillna({'num-of-doors':df['num-of-doors'].mode()[0]})
```

# カテゴリカル変数

- 値が2値
  - label encoding
- 値が3値以上
  - One-Hot Encoding

```python
# カテゴリカル変数を含んだデータのみを抽出
df_obj = df.select_dtypes(include='object')
df_obj.head(3)
```

```python
# nunique()関数でそれぞれの列のユニークな値の数を確認することが出来ます。
df_uni = df_obj.nunique()
df_uni
```

```python
df_cols = df_uni[df_uni==2].index
df_cols
```

```python
from sklearn.preprocessing import LabelEncoder
for col in df_cols:
  le = LabelEncoder()
  le.fit(df[col])
  df[col] = le.transform(df[col])
```

```python
# Encodingが完了していることを確認
df[df_cols].head(2)
```

```python
df.shape
```

```python
df = pd.get_dummies(df, drop_first=True)
```

# 外れ値除去

- 3σ法
- ハンペル判定法

```python
# 分布を表示
sns.distplot(df['price']);
```

```python
mu = df['price'].mean() # 平均値
sigma = df['price'].std() # 標準偏差
```

```python
# 3σ法の中身を取得
df3 = df[(mu - 3 * sigma <= df['price']) & (df['price'] <= mu + 3 * sigma)]
```

```python
# オリジナルのデータフレーム
df.shape
```

```python
# 外れ値除去後のデータフレーム
df3.shape
```

```python
# 分布を表示
sns.distplot(df3['price']);
```

```python
# 中央値を算出
median = df['price'].median()
median
```

```python
# absは指定の値を絶対値に変換。np.medianで中央値を算出
MAD = 1.4826 * np.median(abs(df['price']-median))
MAD
```

```python
df4 = df[(median - 3 * MAD <= df['price']) & (df['price'] <= median + 3 * MAD)]
```
